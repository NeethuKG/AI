```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_iris
```

```python


#Iris is actually a dict
#Iris .keys will show all the feature names
iris =load_iris()
iris
```

```python
iris.keys() # Features
```

```python
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
iris_df.head()
```

```python
iris_df["Species"]=iris.target
iris_df.head()
```

```python
#split the X and y
X= iris_df.drop(['Species'],axis=1)
y=iris_df['Species']
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=17)
print(f"The dimension of X_train is ".format(X_train.shape)")
print(f"The dimension of X_train is ".format(X_test.shape)")
```

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)
```

```python
X_train_score=dt_classifier.score(X_train,y_train)
X_train_score
```

```python
X_test_score=dt_classifier.score(X_test,y_test)
X_test_score
```

```python
y_pred=dt_classifier.predict(X_test)
```

```python
y_pred
```

```python
mse=mean_squared_error(y_test,y_pred)
print("MSE:",mse)
print("RMSE:",mse**(1/2.0))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True)
```

```python
dt_classifier = DecisionTreeClassifier(criterion="entropy",max_depth=3)
dt_classifier.fit(X_train,y_train)
X_train_score=dt_classifier.score(X_train,y_train)
X_train_score
X_test_score=dt_classifier.score(X_test,y_test)
X_test_score
y_pred=dt_classifier.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print("R-Squared Train:",X_train_score)
print("R-Squared Train:",X_test_score)
print("MSE:",mse)
print("RMSE:",mse**(1/2.0))
```
