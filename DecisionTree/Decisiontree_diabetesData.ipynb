```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv("diabetes.csv")
data.head()
```

```python
X=data.drop(["Outcome"],axis=1)
y=data["Outcome"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)
X_train_score=dt_classifier.score(X_train,y_train)

X_test_score=dt_classifier.score(X_test,y_test)
X_test_score
y_pred=dt_classifier.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print("X_train_score:",X_train_score)
print("X_test_score:",X_test_score)

print("MSE:",mse)
print("RMSE:",mse**(1/2.0))

```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True,class_names=True)
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(criterion="entropy",max_depth=3)
dt_classifier.fit(X_train,y_train)
X_train_score=dt_classifier.score(X_train,y_train)
X_train_score
X_test_score=dt_classifier.score(X_test,y_test)
X_test_score
y_pred=dt_classifier.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print("R-Squared Train:",X_train_score)
print("R-Squared Train:",X_test_score)
print("MSE:",mse)
print("RMSE:",mse**(1/2.0))
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True,class_names=True)
plt.show()
```
